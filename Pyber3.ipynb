```python
#Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
city = "city_data1.csv"
ride = "ride_data1.csv"

city_data = pd.read_csv(city, low_memory = False)
ride_data = pd.read_csv(ride,  low_memory = False)

city_data.head()
ride_data.head()
```

```python
city_ride_data = pd.merge(ride_data, city_data, how='left', on=['city', 'city'])
city_ride_data.head()
```

```python
urban_cities = city_ride_data[city_ride_data["type"] == "Urban"]
urban_cities.head()
```

```python
suburban_cities = city_ride_data[city_ride_data["type"] == "Suburban"]
suburban_cities.head()
```

```python
rural_cities = city_ride_data[city_ride_data["type"] == "Rural"]
rural_cities.head()
```

```python
urban_group = urban_cities.groupby("city")
urban_group.head()

urban_average_fare = urban_group["fare"].mean()
urban_average_fare

urban_rides_count = urban_group["fare"].count()
urban_rides_count

urban_driver_count = urban_group["driver_count"].mean()
urban_driver_count
```

```python
suburban_group = suburban_cities.groupby("city")
suburban_group.head()

suburban_average_fare = suburban_group["fare"].mean()
suburban_average_fare

suburban_rides_count = suburban_group["fare"].count()
suburban_rides_count

suburban_driver_count = suburban_group["driver_count"].mean()
suburban_driver_count
```

```python
rural_group = rural_cities.groupby("city")
rural_group.head()

rural_average_fare = rural_group["fare"].mean()
rural_average_fare

rural_rides_count = rural_group["fare"].count()
rural_rides_count

rural_driver_count = rural_group["driver_count"].mean()
rural_driver_count
```

```python
plt.scatter(x=urban_rides_count,
            y=urban_average_fare,
            s=urban_driver_count,
            marker='o', facecolors='red', edgecolors="black",
            alpha=0.9)

plt.scatter(x=suburban_rides_count,
            y=suburban_average_fare,
            s=suburban_driver_count,
            marker='o', facecolors='blue', edgecolors="black",
            alpha=0.9)

plt.scatter(x=rural_rides_count,
            y=rural_average_fare,
            s=rural_driver_count,
            marker='o', facecolors='yellow', edgecolors="black",
            alpha=0.9)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides per city")
plt.ylabel("Average Fare ($) per city")

plt.show()
```

```python
#
#Total Fares Per City Type
#
tot_fares_by_city_type_group = city_ride_data.groupby("type")
percent_total_fares_by_city_type = (tot_fares_by_city_type_group["fare"].sum() / city_ride_data["fare"].sum()) * 100
percent_total_fares_by_city_type

tot_fares_by_city_type_summary = pd.DataFrame ({"Percentage" : percent_total_fares_by_city_type })
tot_fares_by_city_type_summary = tot_fares_by_city_type_summary.reset_index()
tot_fares_by_city_type_summary.head()

colors = ["yellowgreen", "blue", "lightcoral"]
explode = (0, 0, 0)
plt.pie(tot_fares_by_city_type_summary['Percentage'], explode=explode, labels=tot_fares_by_city_type_summary['type'], colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
#Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Fares By City Type")

plt.show()
```

```python
#
#Total Rides Per City Type
#
tot_rides_by_city_type_group = city_ride_data.groupby("type")
percent_total_rides_by_city_type = (tot_rides_by_city_type_group["ride_id"].count() / city_ride_data["ride_id"].count()) * 100
percent_total_rides_by_city_type

tot_rides_by_city_type_summary = pd.DataFrame ({"Percentage" : percent_total_rides_by_city_type })
tot_rides_by_city_type_summary = tot_rides_by_city_type_summary.reset_index()
tot_rides_by_city_type_summary.head()

colors = ["yellowgreen", "blue", "lightcoral"]
explode = (0, 0, 0)
plt.pie(tot_rides_by_city_type_summary['Percentage'], explode=explode, labels=tot_rides_by_city_type_summary['type'], colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
#Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Rides By City Type")

plt.show()
```

```python
#
#Total Drivers Per City Type
#
tot_drivers_by_city_type_group = city_ride_data.groupby("type")
percent_drivers_by_city_type = (tot_drivers_by_city_type_group["driver_count"].count() / city_ride_data["driver_count"].count()) * 100
percent_drivers_by_city_type

tot_drivers_by_city_type_summary = pd.DataFrame ({"Percentage" : percent_drivers_by_city_type })
tot_drivers_by_city_type_summary = tot_drivers_by_city_type_summary.reset_index()
tot_drivers_by_city_type_summary.head()

colors = ["yellowgreen", "blue", "lightcoral"]
explode = (0, 0, 0)
plt.pie(tot_drivers_by_city_type_summary['Percentage'], explode=explode, labels=tot_drivers_by_city_type_summary['type'], colors=colors,
       autopct="%1.1f%%", shadow=True, startangle=140)
#Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Drivers By City Type")

plt.show()
```
