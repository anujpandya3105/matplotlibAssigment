```python
#Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Take in city and ride data and read it into pandas
city = "city_data.csv"
ride = "ride_data.csv"

ride_df = pd.read_csv(ride,  low_memory = False)
city_df = pd.read_csv(city, low_memory = False)

city_df.head()
```

```python
combined_city_ride_dirty_df = pd.merge(ride_df, city_df, how='outer', on='city')
combined_city_ride_dirty_df
combined_city_ride_df = combined_city_ride_dirty_df[['city', 'date', 'fare', 'ride_id', 'driver_count', 'type']]
combined_city_ride_df_clean = combined_city_ride_df.dropna(how='any')
combined_city_ride_df_clean.head(5)
```

```python
group_city = combined_city_ride_df_clean.groupby('city')
group_city.head(5)
```

```python
av_per_city = group_city['fare'].mean()
av_per_city
```

```python
tot_rides_per_city = group_city['ride_id'].count()
tot_rides_per_city
```

```python
rides_df = pd.DataFrame ({"rider_count" : tot_rides_per_city,
                         "average_fare" : av_per_city})

rides_df.head()
```

```python
#new_city_df = city_df.set_index('city')
#new_city_df.head(2)
new_rides_df = rides_df.reset_index()
new_rides_df.head()
city_df.head()
```

```python
#### WHY DO THESE NOT MERGE !!!!!!!!!!!!!!!!!! #####
###To merge 2 data frames do not merge by index, i had to reset the index of the rides_df
### so i reset the index on the rides data frame and then joined city and rides data frames###
###Now I have one consolidated DF, with the city as key.
rides_and_city_merge_df = pd.merge(new_rides_df, city_df, on='city')
rides_and_city_merge_df.head()
```

```python
#plt.scatter(x_axis, data, marker="o", facecolors="red", edgecolors="black",
#            s=x_axis, alpha=0.75)
#rides_and_city_merge_df.plot(kind="scatter", x="rider_count", y="average_fare", s="driver_count", c="type", alpha=0.5)
#rides_and_city_merge_df.plot(kind="scatter", x="rider_count", y="average_fare",  alpha=0.5)
plt.scatter(x=rides_and_city_merge_df['rider_count'],
            y=rides_and_city_merge_df['average_fare'],
            s=rides_and_city_merge_df['driver_count'],
            marker='o', facecolors='red', edgecolors="black",
            alpha=0.5)

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides per city")
plt.ylabel("Average Fare ($) per city")
plt.show()
```

```python
#Total Fares per city type
   
fares_per_city_type_df = rides_and_city_merge_df[ ['average_fare' , 'type'] ]
fares_per_city_type_df.head()
group_fares_by_city_type = fares_per_city_type_df.groupby('type')
group_fares_by_city_type.head(150)

s = group_fares_by_city_type['average_fare'].sum()
c = group_fares_by_city_type['type'].count()

group_fares_by_city_type_summary = pd.DataFrame ({"Percentage" : (s/c) })
new_group_fares_by_city_type_summary= group_fares_by_city_type_summary.reset_index()
new_group_fares_by_city_type_summary.head()
```

```python
colors = ["yellowgreen", "red", "lightcoral"]
explode = (0.1, 0, 0)
plt.pie(new_group_fares_by_city_type_summary['Percentage'], explode=explode, labels=new_group_fares_by_city_type_summary['type'], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Fares per City Type")

plt.show()
```

```python
#Total Rides per city type
   
rides_per_city_type_df = rides_and_city_merge_df[ ['rider_count' , 'type'] ]
rides_per_city_type_df.head()

group_rides_by_city_type = rides_per_city_type_df.groupby('type')
group_rides_by_city_type.head(5)

s = group_rides_by_city_type['rider_count'].sum()
c = group_rides_by_city_type['type'].count()

group_rides_by_city_type_summary = pd.DataFrame ({"Percentage" : (s/c) })
new_group_rides_by_city_type_summary= group_rides_by_city_type_summary.reset_index()
new_group_rides_by_city_type_summary.head()
```

```python
colors = ["yellowgreen", "red", "lightcoral"]
explode = (0.1, 0, 0)
plt.pie(new_group_rides_by_city_type_summary['Percentage'], explode=explode, labels=new_group_rides_by_city_type_summary['type'], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Rides per city type")

plt.show()
```

```python
#
#Total Drivers per city type
#

drivers_per_city_type_df = rides_and_city_merge_df[ ['driver_count' , 'type'] ]
drivers_per_city_type_df.head()

group_drivers_by_city_type = drivers_per_city_type_df.groupby('type')
group_drivers_by_city_type.head(2)

s = group_drivers_by_city_type['driver_count'].sum()
c = group_drivers_by_city_type['type'].count()

group_drivers_by_city_type_summary = pd.DataFrame ({"Percentage" : (s/c) })
new_group_drivers_by_city_type_summary= group_drivers_by_city_type_summary.reset_index()
new_group_drivers_by_city_type_summary.head()


colors = ["yellowgreen", "red", "lightcoral"]
explode = (0.1, 0, 0)
plt.pie(new_group_drivers_by_city_type_summary['Percentage'], explode=explode, labels=new_group_drivers_by_city_type_summary['type'], colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

plt.title("% of Total Drivers By City Type")

plt.show()
```
